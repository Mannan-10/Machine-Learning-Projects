Replacing Categorical data into Numerical data   
Sex : {male : 0,female : 1}    
Smoker : {no : 0, yes : 1}     
Region : {southeast : 0, southwest : 1, northwest : 2, northeast : 3}


```python
data.replace({'sex':{'male':0,'female':1},
              'smoker':{'no':0,'yes':1},
              'region':{'southeast':0,'southwest':1,'northwest':2,'northeast':3}}, inplace=True)
```

```python
data.head()
```

```python
x = data.drop(columns='charges',axis=1)
y = data.charges
```

```python
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
train_data_pred = reg.predict(x_train)
```

```python
r2_train = metrics.r2_score(y_train, train_data_pred)
print("R squared value: ",r2_train)
```

```python
test_data_pred = reg.predict(x_test)
```

```python
r2_test = metrics.r2_score(y_test, test_data_pred)
print("R squared value for test: ",r2_test)
```

Building Predictive system

```python
input_data = (35,0,24.13,1,0,2) # Actual charge is 5125.2157

input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = reg.predict(input_data_reshaped)
print(prediction)

print("The Insurance cost in USD $",prediction[0])
```
