```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit = pd.read_csv('/content/drive/MyDrive/ML Datasets/creditcard.csv')
```

```python
credit.head()
```

```python
credit.tail()
```

```python
credit.shape
```

```python
credit.info()
```

```python
credit.isnull().sum()
```

```python
credit = credit.dropna()
```

```python
credit.isnull().sum()
```

```python
credit.Class.value_counts()
```

```python
credit.Class = credit.Class.astype(int)
```

```python
credit.Class.value_counts()
```

This data is highly unbalanced  
0 -> Normal Transaction   
1 -> Fraud Transaction

```python
legit = credit[credit.Class == 0]
fraud = credit[credit.Class == 1]
```

```python
print(legit.shape,fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=100)
```

```python
data = pd.concat([legit_sample, fraud], axis=0)
```

```python
data.head()
```

```python
data.shape
```

```python
data.groupby('Class').mean()
```

```python
x = data.drop(columns='Class', axis=1)
y = data.Class
```

```python
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_pred = model.predict(x_train)
train_acc = accuracy_score(x_train_pred, y_train)
```

```python
print("Accuracy on Train data: ",train_acc)
```

```python
x_test_pred = model.predict(x_test)
test_acc = accuracy_score(x_test_pred, y_test)
```

```python
print("Accuracy on Test data: ",train_acc)
```
