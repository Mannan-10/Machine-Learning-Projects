```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_data = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_data)
```

```python
house_dataset = pd.DataFrame(house_price_data.data, columns=house_price_data.feature_names)
```

```python
house_dataset.head()
```

```python
house_dataset.shape
```

```python
house_dataset.isnull().sum()
```

```python
house_dataset['price'] = house_price_data.target
```

```python
house_dataset.head()
```

```python
house_dataset.shape
```

```python
house_dataset.isnull().sum()
```

```python
house_dataset.describe()
```

```python
correlation = house_dataset.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Greens')
```

```python
x = house_dataset.drop(['price'], axis=1)
y = house_dataset['price']
```

```python
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction = model.predict(x_train)
```

```python
s1 = metrics.r2_score(y_train, x_train_prediction)
s2 = metrics.mean_absolute_error(y_train, x_train_prediction)

print("R squared error: ",s1)
print("Mean Absolute Error: ",s2)
```

```python
test_prediction = model.predict(x_test)
```

```python
s1 = metrics.r2_score(y_test, test_prediction)
s2 = metrics.mean_absolute_error(y_test, test_prediction)

print("R squared error: ",s1)
print("Mean Absolute Error: ",s2)
```

```python
plt.scatter(y_train, x_train_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
